# Quickstart: AI App

This notebook shows how to use the provider-agnostic LLM client and chat service.

**Prerequisites**: Copy `.env.example` to `.env`, start Ollama, and run from project root (or set `PYTHONPATH` to repo root).

```python
import sys
from pathlib import Path

root = Path.cwd()
if root.name != 'ai-best-practices-example':
    root = root / 'ai-best-practices-example'
if str(root) not in sys.path:
    sys.path.insert(0, str(root))

from config.settings import Settings
from ai_app.llm.openai_compat import OpenAICompatClient
from ai_app.services.chat import ChatService
import asyncio
```

```python
settings = Settings()
client = OpenAICompatClient(
    base_url=settings.llm_base_url,
    api_key=settings.openai_api_key,
    model=settings.llm_model,
    default_max_tokens=100,
)
service = ChatService(client, settings.prompts_dir)

async def run():
    return await service.chat('What is 2+2? Answer in one short sentence.')

reply = asyncio.run(run())
print('Reply:', reply)
```
